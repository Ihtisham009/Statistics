```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_csv('heart.csv')

# 1. Basic Statistical Analysis
summary_stats = data.describe()
correlation_matrix = data.corr()
```

```python

def plot_sex_distribution(data):
    plt.figure(figsize=(6, 4))
    ax = sns.countplot(x='sex', data=data, palette='pastel')
    plt.title('Distribution of Gender')
    plt.xlabel('Gender (0 = Female, 1 = Male)')
    plt.ylabel('Count')
    # Add exact count labels
    for p in ax.patches:
        ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='center', xytext=(0, 5), textcoords='offset points')
    plt.savefig('gender_distribution.png')
    plt.show()


# Function to plot a scatter plot for age vs cholesterol
def plot_age_cholesterol_scatter(data):
    plt.figure(figsize=(8, 5))
    sns.scatterplot(x='age', y='chol', data=data, hue='target', palette='viridis')
    plt.title('Age vs Cholesterol Levels')
    plt.xlabel('Age')
    plt.ylabel('Cholesterol')
    plt.legend(title='Heart Disease (1=Yes, 0=No)')
    plt.savefig('age_cholesterol_scatter.png')  
    plt.show()

# Function to plot a heatmap of correlation matrix
def plot_correlation_heatmap(data):
    plt.figure(figsize=(10, 8))
    sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt='.2f')
    plt.title('Feature Correlation Heatmap')
    plt.savefig('correlation_heatmap.png')  
    plt.show()
    
features = ['thalach']

# Function to create bean plots for each feature in relation to target
def plot_bean_features_by_target(data, features):
    for feature in features:
        plt.figure(figsize=(8, 5))
        sns.violinplot(x='target', y=feature, data=data, palette='Set2')
        plt.title(f'Bean Plot of {feature} by Heart Disease Target')
        plt.xlabel('Heart Disease (1 = Yes, 0 = No)')
        plt.ylabel(feature)
        plt.savefig(f'bean_plot_{feature}_target.png') 
        plt.show()
```

```python

# Generate and display the plots
plot_sex_distribution(data)
plot_age_cholesterol_scatter(data)
plot_correlation_heatmap(data)
plot_bean_features_by_target(data, features)

# Print summaries for inclusion in the report
print("Summary Statistics:\n", summary_stats)
print("\nCorrelation Matrix:\n", correlation_matrix)
```

```python

# Save results to an Excel file
with pd.ExcelWriter('heart_disease_analysis.xlsx') as writer:
    summary_stats.to_excel(writer, sheet_name='Summary Statistics')
    correlation_matrix.to_excel(writer, sheet_name='Correlation Matrix')
print("Results saved to 'heart_disease_analysis.xlsx'")
```
